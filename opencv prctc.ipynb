```python
import cv2 #BGR
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE) #GRAYSCALE--> 0 ,COLOR-->1 , UNCHANGED -->-1
```

```python
cv2.imshow('image',img) #window_name, image
cv2.waitKey(0) #waits for any key to be pressed
cv2.destroyAllWindows() #once the key is pressed it will destroy all windows 
```

```python
plt.imshow(img,cmap='gray',interpolation='bicubic')
plt.plot([50,100],[80,100],'c',linewidth=5) #to draw a line
plt.show()
```

```python
cv2.imwrite('book1.jpeg',img)
```

```python
cap = cv2.VideoCapture(0) #0 means your system 1st webcam , 1 will be for 2nd webcam #instead of zero we can load a video file by giving its name like 'ouput.avi'
# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc,20.0,(640,480))
while True:
    ret,frame = cap.read() #True or False , if there is a feed it returns True
    '''"Frame" will get the next frame in the camera (via "cap"). 
"Ret" will obtain return value from getting the camera frame, either true of false.'''
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) #cvtColor ->convert color and BGR-->Gray
    
    cv2.imshow('frame',frame) #'frame' is just a name of frame
    cv2.imshow('frame2',gray) 
    out.write(frame)
    if cv2.waitKey(1) & 0XFF == ord('q'): #if key q is pressed loop will break 
        '''ord('q') returns the Unicode code point of q
        cv2.waitkey(1) returns a 32-bit integer corresponding to the pressed key
        & 0xFF is a bit mask which sets the left 24 bits to zero, because ord() returns a value betwen 0 and 255, 
        since your keyboard only has a limited character set
Therefore, once the mask is applied, it is then possible to check if it is the corresponding key.'''
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```

# COLORS IN OPENCV
It works on BGR ....
Blue (255,0,0).....
Green (0,255,0)......
Red(0,0,255)......
White(255,255,255)...
Black(0,0,0)

```python
#DRAWING AND WRITING ON IMAGE
img = cv2.imread("book.jpg",cv2.IMREAD_COLOR)
cv2.line(img,(200,200),(300,300),(255,0,0),15) #cv.Line(img, pt1, pt2, color, thickness=15)
cv2.rectangle(img,(15,25),(200,150),(0,0,255),5)  #cv.Rectangle(img, pt1, pt2, color, thickness=1)..pt1 – Vertex of the rectangle...pt2 – Vertex of the rectangle opposite to pt1 
cv2.circle(img,(100,63),55,(0,255,0),-1) #cv2.circle(img,centre of circle,radius,color) , -1 will fill the circle
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,"BOOKS",(0,130),font,1,(0,0,0),2,cv2.LINE_AA) #img,text,start pt,font,size,color,thickness(px),LINE_AA antialiased line
'''In computer graphics, antialiasing is a software technique for diminishing jaggies - stairstep-like 
lines that should be smooth. Jaggies occur because the output device, the monitor or printer,
doesn't have a high enough resolution to represent a smooth line.'''
cv2.imshow("image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Basic Image Operations


```python
#To find the color of a particular pixel
img = cv2.imread("book.jpg",cv2.IMREAD_COLOR)
px = img[55,55] #location
print(px) #actual color value of that pixel
```

```python
#we can modify it to
img[55,55] = [0,0,0]
px = img[55,55]
print(px)
```

```python
# ROI --> Region of image (sub-image of image)
roi = img[100:150,100:150]
print(roi) #printing region
img[100:150,100:150] = [255,255,255] #changing the value of that region to white, that region will appear black
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#changing particular region with another one
change = img[37:111,107:194] #111-37=74 , 194:107=87
img[0:74,0:87] = change
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
